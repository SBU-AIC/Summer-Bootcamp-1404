```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# %matplotlib inline
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y)
plt.title("A Simple Sine Wave")
```

```python
# Create a figure and an axes
fig, ax = plt.subplots()

# Generate some data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Plot the data on the axes
ax.plot(x, y)

# Set labels and title
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title('A Simple Sine Wave')
```

```python
x = np.linspace(0, 2*np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Create the plot
fig, ax = plt.subplots(figsize=(6, 3))
ax.plot(x, y1, label='sin(x)', )
ax.plot(x, y2, label='cos(x)')

# Add labels and title
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title('Sine and Cosine Functions')

# Add a legend
ax.legend()
```

```python
np.random.seed(42)
x = np.random.rand(50)
y = np.random.rand(50)

# Create the plot
fig, ax = plt.subplots(figsize=(6, 4))
ax.scatter(x, y, color='red', alpha=0.5)

# Add labels and title
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title('Random Scatter Plot')
```

```python
np.random.seed(42)
x = np.random.randn(500)
y = np.random.randn(500)

# Create the plot
fig, ax = plt.subplots(figsize=(6, 4))
ax.scatter(x, y, color='red', alpha=0.5)

# Add labels and title
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title('Random Scatter Plot')
```

```python
fig, axs = plt.subplots(2, 2, figsize=(8, 4))
```

```python
x = np.linspace(0, 10, 100)

axs[0, 0].plot(x, np.sin(x))
axs[0, 1].plot(x, np.cos(x))
axs[1, 0].plot(x, np.exp(x))
axs[1, 1].plot(x, np.log(x))

axs[0, 0].set_title('Sine')
axs[0, 1].set_title('Cosine')
axs[1, 0].set_title('Exponential')
axs[1, 1].set_title('Logarithm')

for ax in axs.flat:
    ax.set(xlabel='x-axis', ylabel='y-axis')
    ax.grid(True)

fig
```

```python
np.random.seed(42)
x = np.random.rand(50)
y = x + np.random.normal(0, 0.1, 50)

fig, ax = plt.subplots(figsize=(6, 3))

ax.scatter(x, y)

ax.set_title("Basic Scatter Plot")
ax.set_xlabel("X axis")
ax.set_ylabel("Y axis")
```

```python
categories = ['A', 'B', 'C', 'D', 'E']
values = [23, 35, 14, 27, 10]

fig, ax = plt.subplots(figsize=(7, 5))

ax.bar(categories, values)

ax.set_title("Basic Bar Chart")
ax.set_xlabel("Categories")
ax.set_ylabel("Values")

for i, v in enumerate(values):
    ax.text(i, v, str(v), ha='center', va='bottom')
```
